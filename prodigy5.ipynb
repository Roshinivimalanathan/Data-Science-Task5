```python
data.drop(['Service_year_of_vehicle','Defect_of_vehicle','Work_of_casuality', 'Fitness_of_casuality','Time'],
        axis = 1, inplace = True)
data.head()
```

```python
categorical=[i for i in data.columns if data[i].dtype=='O']
print('The categorical variables are',categorical)
```

```python
for i in categorical:
    data[i].fillna(data[i].mode()[0],inplace=True)
```

```python
data.isna().sum()
```

```python
plt.figure(figsize=(20,200))
plotnumber = 1

for col in categorical:
    if plotnumber <= data.shape[1] and col!='Pedestrian_movement':
        ax1 = plt.subplot(28,1,plotnumber)
        sns.countplot(data, y=col, palette='pastel')
        plt.xticks(fontsize=12)
        plt.yticks(fontsize=12)
        plt.title(col.title(), fontsize=14)
        plt.xlabel('')
        plt.ylabel('')
    plotnumber +=1
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()

#creating a new data frame from performing the chi2 analysis
data1=pd.DataFrame()

#adding all the categorical columns except the output to new data frame
for i in categorical:
    
        data1[i]=le.fit_transform(data[i])
```

```python

x=data1.drop(['Accident_severity'],axis=1)
x.shape
```

```python
x.head
```

```python

y=data1.iloc[:,2]
y.head()
```

```python
sns.countplot(x = y, palette='muted')
```

```python
from imblearn.over_sampling import SMOTE
oversample=SMOTE()
xo,yo=oversample.fit_resample(x,y)
```

```python
from imblearn.over_sampling import SMOTE
oversample=SMOTE()
xo,yo=oversample.fit_resample(x,y)
```

```python
sns.countplot(x = yo, palette='dark')
```

```python
from sklearn.model_selection import train_test_split
#splitting 70% of the data to training data and 30% of data to testing data
x_train,x_test,y_train,y_test=train_test_split(xo,yo,test_size=0.30,random_state=42)
```

```python
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```python
#KNN model alg
from sklearn.neighbors import KNeighborsClassifier
model_KNN=KNeighborsClassifier(n_neighbors=5)
model_KNN.fit(x_train,y_train)
```

```python
y_pred=model_KNN.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score,ConfusionMatrixDisplay
```

```python

report_KNN=classification_report(y_test,y_pred)
print(report_KNN)
```

```python
accuracy_KNN=accuracy_score(y_test,y_pred)
print(accuracy_KNN)
```

```python
matrix_KNN=confusion_matrix(y_test,y_pred)
print(matrix_KNN,'\n')
print(ConfusionMatrixDisplay.from_predictions(y_test,y_pred))
```
